```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import random
import numpy as np
from heapq import nsmallest
from geneticalgorithm import geneticalgorithm as ga
from sklearn.metrics import mean_squared_error 
```

```python
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```python
def plot_data_and_polynomial(x, y, coefficients, m, title):
    x_fit = np.linspace(min(x), max(x), 100)
    y_fit = np.polyval(coefficients, x_fit)

    plt.figure()
    plt.plot(x, y, 'ro', label='Noisy Data')
    plt.plot(x_fit, y_fit, label=title)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
coefficients = np.polyfit(x.flatten(), y.flatten(), 8)
plot_data_and_polynomial(x, y, coefficients, m, f'Polyfit (m={8})')
```

```python
def polynomial_fitness(coeffs):
    poly = np.poly1d(coeffs)
    y_pred = poly(x)
    mse = np.mean((y - y_pred)**2)
    return mse

coeff_bounds = np.array([[-10, 10]] * 8)

model = ga(function=polynomial_fitness, dimension=8, variable_type='real', variable_boundaries=coeff_bounds)
model.run()
```

```python
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)

res = model.best_variable

best_poly = np.poly1d(res)

x_fit = np.linspace(min(x), max(x), 100)
y_fit = np.polyval(coefficients, x_fit)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='green', label='Data')
plt.plot(x_fit, y_fit, color='red', label='Best Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Fitting with Genetic Algorithm')
plt.legend()
plt.grid(True)
plt.show()
```
