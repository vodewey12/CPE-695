### Part 1

````python
import matplotlib.pyplot as plt
import numpy as np

def createXY(noise_scale, number_of_samples):
    x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
    y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    return x, y
def plot_data_and_polynomial(x, y, coefficients, m, title):
    x_fit = np.linspace(min(x), max(x), 100)
    y_fit = np.polyval(coefficients, x_fit)

    plt.figure()
    plt.plot(x, y, 'ro', label='Noisy Data')
    plt.plot(x_fit, y_fit, label=f'Polyfit (m={m})')
    plt.title(title)
    plt.legend()
    plt.show()
    
noise_scale = 100
number_of_samples = 50
x, y = createXY(noise_scale, number_of_samples)

degrees = [2, 3, 4, 5, 6]
for m in degrees:
    coefficients = np.polyfit(x.flatten(), y.flatten(), m)
    plot_data_and_polynomial(x, y, coefficients, m, f'Polyfit (m={m})')
````

### Part 2

````python
from sklearn.metrics import mean_squared_error 

mseArr = []
bestM = 0
lowestMSE = float("inf")
degrees = [i for i in range(1,9)]
for m in degrees:
    coefficients = np.polyfit(x.flatten(), y.flatten(), m)
    y_pred = np.polyval(coefficients, x.flatten())
    mse_value = mean_squared_error(y,y_pred)
    mseArr.append(mse_value)
    if mse_value < lowestMSE:
        bestM = m
        lowestMSE = mse_value
    

plt.figure()
plt.plot(degrees, mse, marker='o')
plt.title('MSE versus Degree of Polynomial')
plt.xlabel('Degree of Polynomial (m)')
plt.ylabel('Mean Squared Error (MSE)')
plt.show()

print(f'The best m based on the MSE is {bestM}. This is because at this degree, the MSE is at the lowest')
````

### Part 3

````python
noise_scales = [150, 200, 400, 600, 1000]

for noise in noise_scales:
    x, y = createXY(noise, number_of_samples)
    coefficients = np.polyfit(x.flatten(), y.flatten(), bestM)
    plot_data_and_polynomial(x, y, coefficients, m, f'Polyfit (m={bestM})')
````

As we can seem, increasing the noise scale directly affects the amount of random data. This make it challenging for the polynomial to fit accurately to the data. Furthermore, as we can see the final graph, the polynomial tries to fit with the noise data, which is inaccurate and not what we want.
